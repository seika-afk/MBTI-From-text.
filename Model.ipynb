```python
porter_stemmer= PorterStemmer()
```

```python
def stemming(content):
    st_content= re.sub('[^a-zA-Z]',' ',content)
    st_content= st_content.lower()
    st_content= st_content.split()
    st_content= [porter_stemmer.stem(word) for word in st_content if not word in stopwords.words('english')]
    st_content= ' '.join(st_content)
    return st_content
```

```python
df["posts"]= df['posts'].apply(stemming)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform

# 1. Define the model
model = LogisticRegression()

# 2. Define the parameter distribution
param_dist = {
    'C': uniform(0.1, 10),  # Uniform distribution between 0.1 and 10
    'solver': ['liblinear', 'saga'],
    'max_iter': [100, 200, 500],
}

# 3. Apply RandomizedSearchCV
random_search = RandomizedSearchCV(estimator=model, param_distributions=param_dist, n_iter=100, cv=5, n_jobs=-1)
random_search.fit(x_train, y_train)

# Get the best parameters
print(f"Best Parameters: {random_search.best_params_}")
print(f"Best Score: {random_search.best_score_}")
```
